## a)

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

x, group = make_blobs(n_samples=1000, centers=2, n_features=4, random_state=0)

# scatter plot for x_1 and x_2
for i in range(len(x)):
    if group[i] == 0:
        plt.scatter(x[i,1], x[i,2], c='steelblue', alpha=0.5)
    else:
        plt.scatter(x[i,1], x[i,2], c='tomato', alpha=0.5)

plt.xlabel('x[:][1]')
plt.ylabel('x[:][2]')

plt.show()
```

## b)

```python
from sklearn.decomposition import PCA
import numpy.linalg as alg


pca = PCA(n_components=4)
x_pca = pca.fit_transform(x)

x_pca_cov = pca.get_covariance()
#print(x_pca_cov)

values, vectors = alg.eigh(x_pca_cov)
print('Covariance matrix: \n',x_pca_cov,'\n')
print('Eigenvalues: ',values)
```

The eigenvalue $\lambda_4=17,52$ is way larger than the other eigenvalues. The eigenvalues are equal to the variances of the succesive principal components. This means that the values of the 4th prinical component have the greatest statistical independence and are therefore topic of the most interest.

## c)

```python
# scatter plot for x_pca_1 and x_pca_2
for i in range(len(x)):
    if group[i] == 0:
        plt.scatter(x_pca[i,1], x_pca[i,2], c='steelblue', alpha=0.5)
    else:
        plt.scatter(x_pca[i,1], x_pca[i,2], c='tomato', alpha=0.5)

plt.xlabel('x_pca[:][1]')
plt.ylabel('x_pca[:][2]')

plt.show()
```

```python
#histograms
plt.rcParams['figure.figsize'] = [15, 10] #larger plots

plot = plt.figure()
ax1 = plot.add_subplot(221,xlabel='x_pca[:][1]', title='PCA X_1')
ax2 = plot.add_subplot(222,xlabel='x_pca[:][2]', title='PCA X_2')
ax3 = plot.add_subplot(223,xlabel='x_pca[:][3]', title='PCA X_3')
ax4 = plot.add_subplot(224,xlabel='x_pca[:][4]', title='PCA X_4')

plt.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.hist(x_pca[:,0], bins=100,color='steelblue')
ax2.hist(x_pca[:,1], bins=100,color='steelblue')
ax3.hist(x_pca[:,2], bins=100,color='steelblue')
ax4.hist(x_pca[:,3], bins=100,color='steelblue')
plt.show()
```
